```python
le = LabelEncoder()

# Fit and transform the 'bought' column
df['bought'] = le.fit_transform(df['bought'])

# Restore the original values using inverse_transform
df['bought'] = le.inverse_transform(df['bought'])

# Display the DataFrame and the type of 'bought'
print(df)
print("Type of 'bought':", df['bought'].dtype)
```
