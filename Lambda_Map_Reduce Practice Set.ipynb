###### Using Lambda Function create a function which can give either even no or odd no for single out put and multiple 
###### output as well also.

```python
#menthod1: single input 
e_o= lambda a: 'even' if a%2==0 else 'odd'
e_o(98)
```

```python
#menthod1: multiple input
lst=[78,73,2,3,4,5,32,43,45,54,12,34,50,83]

def o_e(x):
    if x%2==0:
        return 'even'
    else:
        return 'odd'
```

###### Write a Python program to triple all numbers of a given list of integers. Use Python map
nums1 = [6, 5, 3, 9]

nums2 = [0, 1, 7, 7]

```python
nums1 = [6, 5, 3, 9]
nums2 = [0, 1, 7, 7]

triple_num1= list(map((lambda x: x**3),nums1))
triple_num2= list(map((lambda x: x**3),nums2))

print('Num 1 triple no',triple_num1)
print ('Num 2 triple no', triple_num2)
```

###### Write a Python program to add three given lists using Python map and lambda

```python
nums1 = [6, 5, 3, 9]
nums2 = [0, 1, 7, 7]
nums3=  [9 , 8, 2, 6]
sum_of_all_list= list(map((lambda x,y,z: x+y+z)  , nums1,nums2,nums3))

print('sum_of_all_list', sum_of_all_list)
```

###### Write a Python program 
to convert all the characters in uppercase and lowercase and eliminate duplicate letters 
from a given sequence

a = 'this is PYTHONS class'

```python
a = 'this is PYTHONS class'

lowercase_a = ''.join(map(lambda x: x.lower(), a))

uppercase_a = ''.join(map(lambda x: x.upper(), a))

unique_characters = ''.join(set(a))

print('Lowercase >>>', lowercase_a, 
      '\nUppercase >>>', uppercase_a, 
      '\nUnique Characters >>>', unique_characters)

```
